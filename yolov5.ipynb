```python
x = []
y = []
for i in range(0,7614):
    df = pd.read_csv("/Users/chengziming/Desktop/humanCountData/yolov5_results/yolov5_results/" + str(i) + ".csv")
    y.append(len(df['name'] == 'person') + len(df['name'] == 'zebra')*2)
    x.append(i)
```

```python
plt.plot(x,y)
plt.plot()
```

```python
Savitzky_Golay_smooth = scipy.signal.savgol_filter(y,2001,3)
```

```python
plt.plot(x,Savitzky_Golay_smooth)
plt.plot()
```

```python
slide_human_count = []
slide = 0
for i in range(0,len(y)):
    for j in range(i,i + 20):
        if i >= 7314:
            break
        slide += y[j]
    slide_human_count.append(slide)
    slide = 0
```

```python
plt.plot(x,slide_human_count)
plt.show()
Savitzky_Golay_smooth = scipy.signal.savgol_filter(slide_human_count,2001,3)
plt.plot(x,Savitzky_Golay_smooth)
plt.show()
```
